```python
X = loan_data[['Married', 'ApplicantIncome', 'Education', 'LoanAmount', 'Credit_History']]
```

```python
y = loan_data[['Loan_Status']]
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
y['Loan_Status'] = labelencoder.fit_transform(y['Loan_Status'])
```

```python
y
```

```python
labelencoder.classes_
```

```python
X
```

```python
X['Education'] = labelencoder.fit_transform(X['Education'])
```

```python
X['Married'] = labelencoder.fit_transform(X['Married'])
```

```python
X
```

```python
labelencoder.classes_
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
import joblib 
joblib.dump(scaler,'scaler.pkl')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformance(predictions):
    print("The accuracy score of the model is {}". format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
```

```python
log_predictions = log_model.predict(X_test)
```

```python
modelperformance(log_predictions)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
param_grid = {
    "n_neighbors":[7,9,10,11,12],
    "weights":['uniform','distance']
              }

gridknn = GridSearchCV(knn,param_grid)
gridknn.fit(X_train,y_train)
```

```python
gridknn.best_params_
```

```python
knn_predictions = gridknn.predict(X_test)
```

```python
modelperformance(knn_predictions)
```

```python
from sklearn.svm import SVC
svm = SVC()
param_grid_svc = {
    'C':[0.01,0.05,0.1,0.2,0.5],
    'kernel':['linear','rbf','poly']
    }
gridsvc = GridSearchCV(svm, param_grid_svc)
```

```python
gridsvc.fit(X_train,y_train)
```

```python
gridsvc.best_params_
```

```python
svc_predictions = gridsvc.predict(X_test)
```

```python
modelperformance(svc_predictions)
```

```python
joblib.dump(gridsvc,'model.pkl')
```
