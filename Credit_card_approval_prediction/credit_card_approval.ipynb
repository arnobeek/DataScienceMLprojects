```python
X = data[['Car_Owner','Propert_Owner','Annual_income','EDUCATION']]
y = data[['label']]
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
labelencoder = LabelEncoder()
scaler = StandardScaler()
```

```python
X
```

```python
X['Car_Owner'] = labelencoder.fit_transform(X['Car_Owner'])
```

```python
X['Car_Owner']
```

```python
labelencoder.classes_
```

```python
X['Propert_Owner'] = labelencoder.fit_transform(X['Propert_Owner'])
```

```python
X['Propert_Owner']
```

```python
X['EDUCATION'] = labelencoder.fit_transform(X['EDUCATION'])
```

```python
X['EDUCATION']
```

```python
labelencoder.classes_
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score

def modelperformance(predictions):
    print("The accuracy score of the model is {}".format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_X_train,y_train)
```

```python
log_predictions = log_model.predict(scaled_X_test)
```

```python
modelperformance(log_predictions)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn =KNeighborsClassifier()
param_grid = {
    "n_neighbors":[3,5,7,9,11],
    "weights":["uniform","distance"],
    "algorithm":["auto","kde_tree"]
}
```

```python
grid_knn = GridSearchCV(knn, param_grid)
grid_knn.fit(scaled_X_train,y_train)
```

```python
grid_knn_predictions = grid_knn.predict(scaled_X_test)
```

```python
modelperformance(grid_knn_predictions)
```

```python
grid_knn.best_params_
```

```python
from sklearn.svm import SVC
svm = SVC()

param_grid_svc = {
    "C":[0.01,0.1,0.5,1],
    "kernel":['linear','poly','rbf']
} 
gridsvm = GridSearchCV(svm, param_grid_svc)
gridsvm.fit(scaled_X_train,y_train)
```

```python
gridsvm.best_params_
```

```python
svc_predictions = gridsvm.predict(scaled_X_test)
```

```python
modelperformance(svc_predictions)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
param_grid_tree = {
    "criterion":['gini','entropy'],
    "splitter":['best','random'],
    "max_depth":[None,10,20,30,40,50],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}
gridtree = GridSearchCV(tree,param_grid_tree)
gridtree.fit(scaled_X_train,y_train)
```

```python
gridtree.best_params_
```

```python
tree_predictions = gridtree.predict(scaled_X_test)
modelperformance(tree_predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
param_rfc = {
    "n_estimators":[32,64,128,256],
    "max_features":[2,3,4],
    "bootstrap":[True,False],
    "oob_score":[True, False]
}
```

```python
gridrfc = GridSearchCV(rfc_model,param_rfc)
gridrfc.fit(scaled_X_train,y_train)
```

```python
gridrfc.best_params_
```

```python
rfc_predictions = gridrfc.predict(scaled_X_test)
modelperformance(rfc_predictions)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
param_ada ={
    'n_estimators':[50,100,200],
    'learning_rate':[0.01,0.1,1,10],
}
gridada = GridSearchCV(ada,param_ada)
gridada.fit(scaled_X_train,y_train)
```

```python
ada_predictions = gridada.predict(scaled_X_test)
```

```python
modelperformance(ada_predictions)
```

```python
from xgboost import XGBClassifier
xgb_params = {
    "n_estimators":[20,40,60,80],
    "learning_rate":[0.01,0.1,0.5,1,10],
    "max_depth":[4,6,7,9]
}
gridxgb = GridSearchCV(XGBClassifier(),xgb_params)
gridxgb.fit(scaled_X_train,y_train)
```

```python
gridxgb.best_params_
```

```python
xgb_predictions = gridxgb.predict(scaled_X_test)
modelperformance(xgb_predictions)
```
