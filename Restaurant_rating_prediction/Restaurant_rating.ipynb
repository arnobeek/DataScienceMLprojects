```python
X = restaurant_data[['Average Cost for two','Has Table booking','Has Online delivery', 'Price range']]
```

```python
y = restaurant_data[['Aggregate rating']]
```

```python
X
```

```python
X['Price range'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
X['Has Table booking'] = labelencoder.fit_transform(X['Has Table booking'])
```

```python
X
```

```python
X['Has Online delivery'] = labelencoder.fit_transform(X['Has Online delivery'])
```

```python
labelencoder.classes_
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error,root_mean_squared_error
import numpy as np
```

```python
def modelresults(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(y_test, predictions)))
    print("Root mean squared error on model is {}".format(root_mean_squared_error(y_test, predictions)))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predictionsfromlr = lr.predict(X_test)
```

```python
modelresults(predictionsfromlr)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C":[0.1,0.5],"kernel":["linear","poly"],"degree":[2]}
```

```python
gridsvr = GridSearchCV(svrmodel,param_gridsvr)
gridsvr.fit(X_train,y_train)
```

```python
predgridsvr = gridsvr.predict(X_test)
```

```python
modelresults(predgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
    "max_depth":[10,30],
    "min_samples_leaf":[1,2],
    "min_samples_split":[2,5]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
gridtree = GridSearchCV(estimator=treemodel,param_grid=param_grid)
```

```python
gridtree.fit(X_train,y_train)
```

```python
treepredictions = gridtree.predict(X_test)
```

```python
modelresults(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {
    "max_depth":[5,15],
    "n_estimators":[2,5,10],
}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train,y_train)
```

```python
rfpredictions = gridrfr.predict(X_test)
```

```python
modelresults(rfpredictions)
```

```python
gridrfr.best_params_
```

```python
gridtree.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_param_grid = {"n_neighbors":[9,11,13,15,17,18,19]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
```

```python
knn_grid_search.fit(X_train,y_train)
```

```python
knn_preds = knn_grid_search.predict(X_test)
```

```python
modelresults(knn_preds)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada =AdaBoostRegressor()
ada_param_grid = {
    "n_estimators":[50,100,200],
    "learning_rate":[0.1,0.5,1]
    }
ada_grid_search = GridSearchCV(ada,ada_param_grid)
```

```python
ada_grid_search.fit(X_train,y_train)
```

```python
adapreds = ada_grid_search.predict(X_test)
```

```python
modelresults(adapreds)
```
